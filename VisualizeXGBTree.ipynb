```python
import xgboost as xgb
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
import numpy as np
import pickle
```

```python
pred = np.array([1,1,0,1,0,0,0,0,0,2,2]).reshape(1,-1)
print(len(pred))
with open('featname_simp.pkl', 'rb') as f:
    features = pickle.load(f)
    
pred = xgb.DMatrix(pred, feature_names=features)
print(pred)
```

```python
model = xgb.Booster()
model2 = xgb.Booster()
model3 = xgb.Booster()

model.load_model('otsalary_xgbmodel.bst')
model2.load_model('otsalary_xgbmodel2.bst')
model3.load_model('otsalary_xgbmodel3.bst')

print(model2.predict(pred))

print("Loaded")
```

```python
rcParams['figure.figsize'] = 160,80
xgb.plot_tree(model, numtrees=1)
plt.show()
```

```python
rcParams['figure.figsize'] = 160,80
xgb.plot_tree(model3, numtrees=5)
plt.show()
```
